# Analyze Processing Quality

#  

Analyze the overall quality of the processing of the first batch of Alouette images. Determine that all subdirectories have been processed. Determine which subdirectories have the highest loss - are there any patterns? For instance, do certain subdirectories have high loss, or images from certain ground stations?

```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
```

```python
rootDir = 'L:/DATA/Alouette_I/BATCH_I_Run1/'
rootDir_md = 'L:/DATA/Alouette_I/BATCH_I_Run1_md/'
resultDir = rootDir + '05_result/'
resultDir_md = rootDir_md + '05_result/'
logDir = rootDir + '06_log/'
logDir_md = rootDir_md + '06_log/'
```

#  

#### Identify outstanding subdirectories to process:

Check from process_log:

Remove duplicates from process_log:

```python
df_process_log = pd.read_csv(logDir + 'process_log.csv')
print(len(df_process_log))
df_process_log.head()
```

```python
df_p1 = df_process_log.sort_values('Process_timestamp', ascending=True)
df_p1 = df_p1.drop_duplicates(subset=['subdir_id'], keep='last') 
print(len(df_p1))
df_p1.head()
```

```python
df_p1.tail()
```

Look at duplicates from process_log:

```python
df_p1_ = df_process_log[df_process_log.duplicated(subset=['subdir_id'], keep=False)]
df_p1_ = df_p1_.sort_values(['subdir_id', 'Process_timestamp'])
print(len(df_p1_))
df_p1_.head(10)
```

Find any subdirectories that have not been processed:

```python
df_inventory = pd.read_csv(logDir + 'image_inventory.csv')
print(len(df_inventory))
df_inventory.head()
```

```python
subdir_ids_tot = df_inventory['subdir_id'].unique()
subdir_ids_proc = df_p1['subdir_id'].unique()
subdir_ids_rem = list(set(subdir_ids_tot) - set(subdir_ids_proc))
subdir_ids_rem
```

Check from resultDir:

```python
df_result = pd.DataFrame()
i = 0
for file in os.listdir(resultDir):
    if 'R' in file:
        directory = file
        for file2 in os.listdir(resultDir + directory + '/'):
            if 'result' in file2:
                fn_parts = file2.split('_')
                subdirectory = fn_parts[1].replace('.csv', '')
                try:
                    df_load = pd.read_csv(resultDir + directory + '/' + file2, sep=',')
                    n = len(df_load)
                except pd.errors.EmptyDataError:
                    n = 0
                row = pd.DataFrame({
                    'Directory': directory,
                    'Subdirectory': subdirectory,
                    'images_result': n,
                    'subdir_id': directory + '_' + subdirectory
                }, index=[i])
                df_result = pd.concat([df_result, row])
                i += 1
                if i % 100 == 0:
                    print(str(i) + ': Now inspecting...' + directory + '/' + subdirectory)
```

```python
print(len(df_result))
df_result.head()
```

```python
subdir_ids_tot = df_inventory['subdir_id'].unique()
subdir_ids_res = df_result['subdir_id'].unique()
subdir_ids_rem = list(set(subdir_ids_tot) - set(subdir_ids_res))
subdir_ids_rem
```

```python
df_p1_res = df_p1.merge(df_result, on=['Directory', 'Subdirectory', 'subdir_id'], how='left')
print(len(df_p1_res))
df_p1_res.head()
```

```python
df1 = df_p1_res.loc[df_p1_res['Images_processed'] == 0]
df2 = df_p1_res.loc[df_p1_res['images_result'] == 0]
df3 = df_p1_res.loc[(df_p1_res['Images_processed'] == 0) | (df_p1_res['images_result'] == 0)]
print(len(df1))
print(len(df2))
print(len(df3))
df3.head()
```

Check mapped_coords:

```python
df_mc = pd.DataFrame()
i = 0
for file in os.listdir(resultDir + 'mapped_coords/'):
    if 'R' in file:
        directory = file
        for file2 in os.listdir(resultDir + 'mapped_coords/' + directory + '/'):
            subdirectory = file2
            n = 0
            for file3 in os.listdir(resultDir + 'mapped_coords/' + directory + '/' + subdirectory + '/'):
                if 'mapped_coords' in file3:
                    n += 1
            row = pd.DataFrame({
                'Directory': directory,
                'Subdirectory': subdirectory,
                'mapped_coords_result': n,
                'subdir_id': directory + '_' + subdirectory
            }, index=[i])
            df_mc = pd.concat([df_mc, row])
            i += 1
            if i % 100 == 0:
                    print(str(i) + ': Now inspecting...' + directory + '/' + subdirectory)
```

```python
print(len(df_mc))
df_mc.head()
```

```python
df_p1_mc = df_p1.merge(df_mc, on=['Directory', 'Subdirectory', 'subdir_id'], how='left')
print(len(df_p1_mc))
df_p1_mc.head()
```

```python
df1 = df_p1_mc.loc[df_p1_mc['Images_processed'] == 0]
df2 = df_p1_mc.loc[df_p1_mc['mapped_coords_result'] == 0]
df2a = df_p1_mc.loc[df_p1_mc['mapped_coords_result'].isnull()]
df3 = df_p1_mc.loc[(df_p1_mc['Images_processed'] == 0) | (df_p1_mc['mapped_coords_result'] == 0)]
print(len(df1))
print(len(df2))
print(len(df2a))
print(len(df3))
df3
```

#  

#### Merge re-processed metadata process_log (BATCH_I_Run1_md) with BATCH_I_Run1 process_log:

```python
df_process_log_md = pd.read_csv(logDir_md + 'process_log.csv')
print(len(df_process_log_md))
df_process_log_md.head()
```

```python
df_p1_md = df_process_log_md.sort_values('Process_timestamp', ascending=True)
df_p1_md = df_p1_md.drop_duplicates(subset=['subdir_id'], keep='last') 
print(len(df_p1_md))
df_p1_md.head()
```

```python
df_p1_md_ = df_process_log_md[df_process_log_md.duplicated(subset=['subdir_id'], keep=False)]
df_p1_md_ = df_p1_md_.sort_values(['subdir_id', 'Process_timestamp'])
print(len(df_p1_md_))
df_p1_md_.head(10)
```

```python
df_p1 = df_p1.add_suffix('_Run1')
df_p1_md = df_p1_md.add_suffix('_Run1_md')
df_p2 = df_p1.merge(df_p1_md, how='left', left_on=['Directory_Run1', 'Subdirectory_Run1', 'subdir_id_Run1'], 
                    right_on=['Directory_Run1_md', 'Subdirectory_Run1_md', 'subdir_id_Run1_md'])
df_p2 = df_p2.rename(columns={
    'Directory_Run1': 'Directory',
    'Subdirectory_Run1': 'Subdirectory',
    'subdir_id_Run1': 'subdir_id'
})
print(len(df_p2))
df_p2.sample(5)
```

Images_processed is the limiting variable. For each subdirectory, find the lower 'Images_processed' value (whether from BATCH_I_Run1 or BATCH_I_Run1_md):

```python
df_p2['Images_processed'] = 0
df_p2['Images_dot'] = 0
df_p2['Images_num'] = 0
df_p2['Images_loss'] = 0
df_p2['Images_outlier'] = 0
for i in range(0, len(df_p2)):
    if df_p2['Images_processed_Run1'].iloc[i] <= df_p2['Images_processed_Run1_md'].iloc[i]:
        suffix = '_Run1'
    else:
        suffix = '_Run1_md'
    df_p2['Images_processed'].iloc[i] = df_p2['Images_processed' + suffix].iloc[i]
    df_p2['Images_dot'].iloc[i] = df_p2['Images_dot' + suffix].iloc[i]
    df_p2['Images_num'].iloc[i] = df_p2['Images_num' + suffix].iloc[i]
    df_p2['Images_loss'].iloc[i] = df_p2['Images_loss' + suffix].iloc[i]
    df_p2['Images_outlier'].iloc[i] = df_p2['Images_outlier' + suffix].iloc[i]
```

```python
print(len(df_p2))
df_p2.head()
```

#  

Calculate percent 'total loss':

```python
df1['Images_total_loss'] = df1['Images_unprocessed'] + df1['Images_loss'] + df1['Images_outlier']
df1['pc_total_loss'] = df1['Images_total_loss']/df1['Images_downloaded']
df1['pc_unprocessed'] = df1['Images_unprocessed']/df1['Images_downloaded']
df1['pc_processed'] = df1['Images_processed']/df1['Images_downloaded']
df1['pc_dot'] = df1['Images_dot']/df1['Images_downloaded']
df1['pc_num'] = df1['Images_num']/df1['Images_downloaded']
df1['pc_loss'] = df1['Images_loss']/df1['Images_downloaded']
df1['pc_outlier'] = df1['Images_outlier']/df1['Images_downloaded']
df2 = df1.sort_values('pc_total_loss', ascending=False).reset_index(drop=True)
df2.head()
```

```python
#Merge BATCH_I_Run1 results with BATCH_I_Run1_md results

df_result = pd.DataFrame()
for file in os.listdir(resultDir):
    if 'R' in file:
        print('Reading...' + file)
        for file2 in os.listdir(resultDir + file):
            if 'result' in file2:
                #Load BATCH_I_Run1 result
                try:
                    df_load = pd.read_csv(resultDir + file + '/' + file2)
                except pd.errors.EmptyDataError:
                    continue
                #Load BATCH_I_Run1_md result
                try:
                    df_load_md = pd.read_csv(resultDir_md + file + '/' + file2)
                    #Merge
                    df_load = df_load.add_suffix('_Run1')
                    df_load_md = df_load_md.add_suffix('_Run1_md')
                    df_merge = df_load.merge(df_load_md, how='left', left_on=['Directory_Run1', 'Subdirectory_Run1', 
                                                                              'filename_Run1'], 
                                        right_on=['Directory_Run1_md', 'Subdirectory_Run1_md', 'filename_Run1_md'])
                    df_merge = df_merge.rename(columns={
                        'Directory_Run1': 'Directory',
                        'Subdirectory_Run1': 'Subdirectory',
                        'filename_Run1': 'filename'
                    })
                    df_result = pd.concat([df_result, df_merge])
                except pd.errors.EmptyDataError:
                    continue
```

```python
print(len(df_result))
df_result.head()
```

```python
df_result.to_csv(resultDir + 'result_total_mdrp.csv', index=False)
```
