```python
resultDir = 'L:/DATA/Alouette_I/Total/05_result/'
dataDir_I = 'L:/DATA/Alouette_I/BATCH_I_Run2/04_processed/'
dataDir_II = 'L:/DATA/Alouette_I/BATCH_II_Run2/04_processed/'
```

```python
df_result = pd.read_csv(resultDir + 'result_master.csv')
print(len(df_result))
df_result.head()
```

```python
df_s2 = pd.read_csv(resultDir + 'result_stage2_raw.csv', low_memory=False)
print(len(df_s2))
df_s2.head()
```

```python
df_s2_num = df_s2.loc[df_s2['processed_image_class'] == 'num']
print(len(df_s2_num))
df_s2_num.head()
```

```python
df_s2_num['satellite_number'].unique()
```

```python
df_s2_num_satnum_not1 = df_s2_num.loc[df_s2_num['satellite_number'] != 1]
print(len(df_s2_num_satnum_not1))
df_s2_num_sam = df_s2_num_satnum_not1.sample(10)
df_s2_num_sam
```

```python
for i in range(0, len(df_s2_num_sam)):
    fn_pathend = df_s2_num_sam['Directory'].iloc[i] + '/' + df_s2_num_sam['Subdirectory'].iloc[i] + '/' + df_s2_num_sam['filename'].iloc[i]
    try:
        img = mpimg.imread(dataDir_I + fn_pathend)
    except:
        img = mpimg.imread(dataDir_II + fn_pathend)

    plt.imshow(img)
    plt.show()
    print(df_s2_num_sam.iloc[i:i+1][['satellite_number', 'year', 'day_1', 'day_2', 'day_3', 'hour_1', 'hour_2', 'minute_1', 'minute_2', 
                   'second_1', 'second_2', 'station_number_1', 'station_number_2', 'processed_image_class']])
```
