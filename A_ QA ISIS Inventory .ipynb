```python
import os
```

## Defining the functions

```python
def Directory_count(folder_paths):
    total_directories = 0
    total_subdirectories = 0

    for folder_path in folder_paths:
        num_directory = 0
        num_subdirectories = 0

        items = os.listdir(folder_path)
        subfolders = [item for item in items if os.path.isdir(os.path.join(folder_path, item))]
        num_directory = len(subfolders)
        total_directories += num_directory

        print(f"Number of Directories in '{folder_path}': {num_directory}")

        for subfolder in subfolders:
            subfolder_path = os.path.join(folder_path, subfolder)
            subfolder_items = os.listdir(subfolder_path)
            num_subfolder = len([item for item in subfolder_items if os.path.isdir(os.path.join(subfolder_path, item))])
            num_subdirectories += num_subfolder

            print(f"  - Number of Subdirectories in '{subfolder}': {num_subfolder}")

        total_subdirectories += num_subdirectories

    print(f"\nTotal Number of Directories: {total_directories}")
    print(f"Total Number of Subdirectories: {total_subdirectories}")
```

```python
def count_png_files(folder):
    png_count = 0

    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.lower().endswith(".png"):
                png_count += 1

    print(f"Total number of PNG files in '{folder}' and its subfolders: {png_count}")
```

```python
def find_folders_without_keywords(root_folders, keywords):
    matching_folders = []
    
    for root_folder in root_folders:
        for root, dirs, files in os.walk(root_folder):
            if all(keyword not in root for keyword in keywords):
                matching_folders.append(root)
    
    return matching_folders

def count_images_in_folder(folder_path):
    image_extensions = [".png"]
    image_count = 0
    
    for file in os.listdir(folder_path):
        if any(file.lower().endswith(ext) for ext in image_extensions):
            image_count += 1
    
    return image_count
```

# Total number of ISIS 1 folders and images (aka ISIS A, ISIS-A) 

```python
root_folder_paths = ['L:/DATA/ISIS/ISIS_101300030772/', 'L:/DATA/ISIS/ISIS_102000056114/', 'L:/DATA/ISIS/raw_upload_20230421']
keywords = ["ISIS A", "ISIS-A"]

total_image_count = 0

for root_folder_path in root_folder_paths:
    matching_folders = find_folders_with_keywords([root_folder_path], keywords)

    if matching_folders:
        print(f"Number of folders containing {', '.join(keywords)} in {root_folder_path}: {len(matching_folders)}")
        print(f"Folders:")
        for folder in matching_folders:
            image_count = count_images_in_folder(folder)
            total_image_count += image_count
            print(f"{folder} - Images: {image_count}")
    else:
        print(f"No folders containing {', '.join(keywords)} found in {root_folder_path}.")
    print()

print(f"Total number of images in all matching folders: {total_image_count}")
```
