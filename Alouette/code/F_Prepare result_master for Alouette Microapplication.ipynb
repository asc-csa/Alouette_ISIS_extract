#  

Shape processed image results data to fit current Alouette microapplication (https://donnees-data.asc-csa.gc.ca:8888/alouette/):

```python
import pandas as pd
```

```python
dataDir = 'L:/DATA/Alouette_I/Total/05_result/'
```

#  

#### Functions:

```python
def fix_Lat(x):
    
    if type(x) == str:
        if 'S' in x:
            x = x.replace('S', '')
            x = float(x)
            x = -1*x
        elif 'N' in x:
            x = x.replace('N', '')

    return x
```

```python
def fix_Lng(x):
    
    if type(x) == str:
        if 'W' in x:
            x = x.replace('W', '')
            x = float(x)
            x = -1*x
        elif 'E' in x:
            x = x.replace('E', '')

    return x
```

#  

#### Shape results to fit microapplication data model:

```python
df_result = pd.read_csv(dataDir + 'result_master.csv')
print(len(df_result))
df_result.head()
```

```python
df_result['ID'] = df_result.index
df_result['File name'] = df_result['Directory'].astype(str) + '/' + df_result['Subdirectory'].astype(str) + '/' + df_result['filename'].str.replace('.png', '')
df_result['Minimum frequency'] = df_result['fmin']
df_result['Maximum depth'] = df_result['max_depth']
df_result['Subdirectory name'] = df_result['Subdirectory']
df_result['Satellite number'] = 1
df_result['Ground station number'] = df_result['Station_Number']
df_result['Ground station name'] = df_result['Station_Name']
df_result['Ground station code'] = df_result['Station_Code']
df_result['Latitude'] = df_result['Lat']
df_result['Longitude'] = df_result['Lng']
```

```python
print(len(df_result))
df_result.head()
```

Adjust Lat, Lng:

```python
df_result['Latitude'] = df_result['Latitude'].fillna('')
df_result['Latitude'] = df_result['Latitude'].apply(lambda x: fix_Lat(x))
```

```python
df_result['Longitude'] = df_result['Longitude'].fillna('')
df_result['Longitude'] = df_result['Longitude'].apply(lambda x: fix_Lng(x))
```

```python
print(len(df_result))
df_result.head()
```

Save:

```python
df_result2 = df_result[['ID', 'File name', 'Minimum frequency', 'Maximum depth', 
                        'Subdirectory name', 'Satellite number', 'Ground station number', 
                        'Timestamp', 'Ground station name', 'Ground station code', 'Latitude', 
                        'Longitude']]
print(len(df_result2))
df_result2.head()
```

```python
df_result2.to_csv(dataDir + 'result_microapp.csv', index=False)
```
